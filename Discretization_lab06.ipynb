## Exercise #11 Discretization

Load mtcar dataset

```python
# Load the mtcars dataset using R's datasets library
%load_ext rpy2.ipython
%R library(datasets)
%R data(mtcars)

# Display the structure of the dataset
%R str(mtcars)

# Display the first few rows of the dataset
%R head(mtcars)
```

```python
# Discretization of 'mpg' variable
%R library(dplyr)
%R mtcars <- mtcars %>% mutate(mpg_category = cut(mpg, breaks=c(0, 15, 20, 35), labels=c("Low", "Medium", "High"), right=FALSE))

# Plot the distribution of 'mpg_category'
%R library(ggplot2)
%R ggplot(mtcars, aes(x=mpg_category)) + geom_bar() + labs(title='Distribution of MPG Categories', x='MPG Category', y='Count')
```

# Apply Discretization on Different Variables

```python
# Import the necessary libraries
import pandas as pd

# Load the mtcars dataset from R into a Pandas DataFrame
%load_ext rpy2.ipython
%R library(datasets)
%R data(mtcars)
%R -o mtcars

# Define the bins and labels for discretization
bins = [0, 15, 20, 35]
labels = ["Low", "Medium", "High"]

# Create a new column 'mpg_category' with discretized values
mtcars['mpg_category'] = pd.cut(mtcars['mpg'], bins=bins, labels=labels, right=False)

# Display the first few rows of the dataset with the new column
mtcars.head()
```

# Plotting the results

```python
import matplotlib.pyplot as plt

# Plot the distribution of 'mpg_category'
plt.figure(figsize=(8, 6))
mtcars['mpg_category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of MPG Categories')
plt.xlabel('MPG Category')
plt.ylabel('Count')
plt.show()
```

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the mtcars dataset from R into a Pandas DataFrame
%load_ext rpy2.ipython
%R library(datasets)
%R data(mtcars)
%R -o mtcars

# Define the bins and labels for discretization
bins = [0, 4, 6, 8]
labels = ["4 Cyl", "6 Cyl", "8 Cyl"]

# Create a new column 'cyl_category' with discretized values
mtcars['cyl_category'] = pd.cut(mtcars['cyl'], bins=bins, labels=labels, right=False)

# Plot the distribution of 'cyl_category'
plt.figure(figsize=(8, 6))
mtcars['cyl_category'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Cylinder Categories')
plt.xlabel('Cylinder Category')
plt.ylabel('Count')
plt.show()
```
