Solution of exercise 06

```
# Load the dataset
data(airquality)

# EDA
str(airquality)
summary(airquality)
head(airquality)
tail(airquality)

# Check for missing values
missing_values <- colSums(is.na(airquality))
missing_values

# Impute missing values with mean
airquality$Ozone[is.na(airquality$Ozone)] <- mean(airquality$Ozone, na.rm = TRUE)
airquality$Solar.R[is.na(airquality$Solar.R)] <- mean(airquality$Solar.R, na.rm = TRUE)

# Apply scaling
scaled_airquality <- airquality

# Divide data into training and testing sets
set.seed(123)
train_indices <- sample(1:nrow(airquality), 0.7 * nrow(airquality))
train_data <- airquality[train_indices, ]
test_data <- airquality[-train_indices, ]

# Build the Linear Regression model
lm_model <- lm(Ozone ~ ., data = train_data)

# Predict using the model
predictions <- predict(lm_model, newdata = test_data)

#  difference between actual and predicted values
differences <- test_data$Ozone - predictions

# Model Output
print(lm_model)

# Model Summary
summary(lm_model)

# Model Visualization (scatterplot of actual vs. predicted values)
plot(test_data$Ozone, predictions, xlab = "Actual Ozone Levels", ylab = "Predicted Ozone Levels", main = "Actual vs. Predicted Ozone Levels")
abline(0, 1, col = "blue")

# Show the difference between actual values and predicted results
cat("Difference between actual and predicted values:\n")
print(differences)
```

## Model Visualization
Above figure is a scatter plot of actual vs predicted ozone levels, which depicts how the model performed. It should be noticed, that every dot on the picture reflects observation. On the x axis is shown Actual Ozone Levels, while on the y axis is presented the Predicted Ozone Levels by the model.

The blue line on the plot is the perfect predictive equation that shows how values in reality equal forecasted values. For perfect prediction, the scatter plot points should lie on this line.

Here, the points scatter a bit around the line, meaning that the model predictions come really close to reality, with certain amount of variability.

Points above the line indicate that the model has overestimated something, while points below the line mean that it has underestimated something.

Scatter plots visually depict how effective the chosen model is in estimating ozone levels, highlighting zones that may require betterment.

```
library(arules)
library(arulesViz)

# Define the URL of the CSV file on GitHub
csv_url <- 'https://raw.githubusercontent.com/itechplus2020/DS2023/main/Telco-Customer-Churn.csv'
```

```
telco_data <- read.csv(url(csv_url))

# Convert categorical columns to factors
factor_columns <- c(
  "gender", "SeniorCitizen", "Partner", "Dependents", "PhoneService",
  "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup",
  "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies",
  "Contract", "PaperlessBilling", "PaymentMethod", "Churn"
)
```

```
telco_data[factor_columns] <- lapply(telco_data[factor_columns], as.factor)

# Convert the dataset to a transaction format
telco_transactions <- as(telco_data, "transactions")
```

```
# Mine association rules with specified support and confidence thresholds
association_rules <- apriori(telco_transactions, parameter = list(support = 0.03, confidence = 0.8))

# Remove redundant rules
association_rules <- sort(association_rules, by = "lift")

# Extract rules with a minimum support of 0.03 and a minimum confidence of 0.8
filtered_rules <- subset(association_rules, support > 0.03 & confidence > 0.8)

# Visualize the results
plot(filtered_rules, method = "graph", control = list(type = "items"))
```
